```python
import requests # get post put patch APIS
client_id = ""
client_secret = ""

url = "https://api.ebay.com/identity/v1/oauth2/token"
data = {
    "grant_type": "client_credentials",
    "scope": "https://api.ebay.com/oauth/api_scope"
}

response= requests.post(url,
                       auth=(client_id,client_secret),
                       data=data)


token = response.json()["access_token"]
print(token)
```

```python
import pandas as pd
search_url = "https://api.ebay.com/buy/browse/v1/item_summary/search" #API browse
y =input("Enter the theme for your product:")
x=int(input("How many items do you want to fetch?"))

all_items=[]
limit=50
total_needed=x

for offset in range(0, total_needed, 50):
    response = requests.get(
        "https://api.ebay.com/buy/browse/v1/item_summary/search",
        params={
            "q": y,
            "limit": 50,
            "offset": offset
        },
        headers={
            "Authorization": f"Bearer {token}"
        }
    )
    data = response.json()
    all_items.extend(data.get('itemSummaries', []))


df=pd.DataFrame(all_items)
df
```

```python
import mysql.connector
from sqlalchemy import create_engine

conn = mysql.connector.connect(
  host="localhost",
  user="root",
  password=""
)
cursor = conn.cursor()
cursor.execute("CREATE DATABASE IF NOT EXISTS depi")
cursor.execute("USE depi")
cursor.execute('''
CREATE TABLE IF NOT EXISTS DimDate (
    DateId INT AUTO_INCREMENT PRIMARY KEY,
    FullDate Date,
    Day int,
  Month INT,
               Year INT,
               Hour INT,
               Minute INT,
               Second INT,
               MonthName VARCHAR(100)
)
''')
user = "root"
password = ""
host = "localhost"
database = "depi"
engine = create_engine(f"mysql+mysqlconnector://{user}:{password}@{host}/{database}")
DimDate.to_sql(
    name='DimDate',   # Table name in MySQL
    con=engine,        # Connection engine
    if_exists='append',  # 'replace' = overwrite table, 'append' = add rows
    index=False        # Do not write DataFrame index as a column
)



conn.commit()
```
